```python
from paddleocr import PaddleOCR, draw_ocr

ocr = PaddleOCR(use_angle_cls = True, lang='en')
```

```python
result = ocr.ocr(image_file, cls=True)
textRecognized = []
for line in result:
  if len(line)>2:
    for i in range(len(line)):
      textRecognized.append(line[i])
for i in range(len(textRecognized)):
  print(textRecognized[i])
```

#### Drawing results

```python
from PIL import Image, ImageDraw, ImageFont
image = Image.open(image_file).convert('RGB')

boxes = [textRecognized_[0] for textRecognized_ in textRecognized]
for box in boxes:
  print(box)
```

```python
txts = [textRecognized_[1][0] for textRecognized_ in textRecognized]
for txt in txts:
  print(txt)
```

```python
scores = [textRecognized_[1][1] for textRecognized_ in textRecognized]
for score in scores:
  print(score)
```

```python
import os

for root, dirs, files in os.walk("/usr/share/fonts/"):
    for file in files:
        if file.endswith(".ttf"):
            print(os.path.join(root, file))
```

```python
font_path = "/usr/share/fonts/truetype/liberation/LiberationSerif-Regular.ttf"

# Draw OCR results using the correct font path
result_img = draw_ocr(image, boxes, txts, scores, font_path=font_path)

img = Image.fromarray(result_img)
img.save('results.jpg')
```
